```python
import pandas as pd

df = pd.read_csv('C:/Users/thous/OneDrive/Desktop/NUH/test.csv', encoding= 'utf-8')

df1 = df.filter(['Enter Room', 'Exit Room'], axis=1)

df1[['Enter Room', 'Entertime']] = df1['Enter Room'].str.split(' ', expand=True)
df1[['Exit Room', 'Exittime']] = df1['Exit Room'].str.split(' ', expand=True)

df1.head()
```

```python
df['Enter Room'] = pd.to_datetime(df['Enter Room'], format = '%d-%m-%Y %H:%M')

df.head()
```

```python
df['Day'] = df['Enter Room'].dt.day_name()
df
```

```python
df['Exit Room'] = pd.to_datetime(df['Exit Room'], format = '%d-%m-%Y %H:%M')
df.head()
```

```python
df['time_diff'] = df['Exit Room']-df['Enter Room']
df.head()
```

```python
df1['Entertime'] = pd.to_datetime(df1['Entertime'], format = '%H:%M')
df1['Exittime'] = pd.to_datetime(df1['Exittime'], format = '%H:%M')

from datetime import datetime

start = datetime.strptime('09:00:00', '%H:%M:%S')
end = datetime.strptime('17:30:00', '%H:%M:%S')

def out(df1):
    if df1['Entertime'] < start: 
        return 'before'
    elif df1['Exittime'] > end:
        return 'after'
    else:
        return 'no'
    
df1['outside hours'] = df1.apply(out, axis=1)

df1.head()
```

```python
df['outside hours'] = df1['outside hours']

df.head()
```

```python
df['time_diff'] = df['time_diff'].astype(str).str[-18:-10]

df.head()
```
